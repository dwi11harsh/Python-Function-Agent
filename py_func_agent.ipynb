# Python Function Agent

The idea is to create an api that will create a python function and also test it and return it only when the function executes and generates the desired output.

## Expected Results
Ideally the agent should be able to do the following:

1. Process the prompt and if the requirements are too vague then generate a proper prompt.  
1.1 if the function size is too big then it needs to be broken down to smaller functions.  
1.2 if the request is for something that cannot be tested properly (like a BE server) then we need to return a warning from this step itself.
2. There needs to be a storage facility from where the execution is done with ease (not sure what to use rn).
3. The agent should have access to a virtual environment and install any libraries that might be required in order to execute the generated function.
4. The agent should then be able to articulate some extra code for proper testing of the function.
5. The agent needs to have access to the output of the function execution.
6. If the function execution completes with a success then this function will be returned to the user.
7. If for some reason there is an error then the llm needs to fix it.
8. The llm will only get three trials to fix this error and if it fails then the agent will take all the generated functions, errors and corrections and try to generate a new function keeping the previous chats in context.
